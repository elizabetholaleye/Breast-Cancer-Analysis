```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Load Breast Cancer dataset
data = load_breast_cancer()
bc_df = pd.DataFrame(data.data, columns=data.feature_names)
bc_df['target'] = data.target

# Plot the distribution of the target variable
plt.figure(figsize=(10, 6))
sns.countplot(x='target', data=bc_df)
plt.title('Distribution of Breast Cancer Diagnosis (Malignant vs. Benign)')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Malignant', 'Benign'])
plt.show()
```

```python
# Boxplot of 'breastcancer' content
plt.figure(figsize=(8, 6))
sns.boxplot(x=bc_df['mean radius'])
plt.title('Boxplot of BC')
plt.xlabel('BC Content (%)')
plt.show()
```

```python
# Bar plot of 'target' (class)
plt.figure(figsize=(10, 6))
sns.countplot(x='target', data=bc_df)
plt.title('Distribution of Breast Cancer Classes')
plt.xlabel('BreastCancer Class')
plt.show()
```

```python
# Create a two-way table
pd.crosstab(bc_df['mean radius'], bc_df['target'])

#This shows how many members of each class have a specific alcohol content
```

```python
# Scatter plot of 'alcohol' content vs 'flavanoids' with target classes
plt.figure(figsize=(10, 6))
sns.scatterplot(x='mean radius', y='mean texture', hue='target', data=bc_df)
plt.title('Mean radius vs Mean texture with Target Classes')
plt.xlabel('mean radius')
plt.ylabel('mean texture')
plt.show()
```
